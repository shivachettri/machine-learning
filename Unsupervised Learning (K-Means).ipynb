### Unsupervised Learning (K-Means)
In upsupervised learning, the dataset has no label and we have to draw interference from dataset of our own.

```python
from sklearn import datasets
from sklearn.cluster import KMeans
```

#### Clustering
Clustering is a method of classifying data into groups using centroids (center point). K-Means is a clustering algorithm that iteratively classifies data in classes.

Next, we we are to give some input then K-Means algorithm is going to seggregate data into respected cluster/class.

We will use default data set from sklearn library called 'iris dataset' for our purpose.
